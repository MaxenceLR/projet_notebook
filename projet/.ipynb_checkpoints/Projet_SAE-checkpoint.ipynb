Creation d'un histogramme montrant la disturbution des salaires en USD avec la biblotheque matplotlib et la methode "hist"

```python
# Distribution des salaires en USD
%matplotlib inline
plt.hist(data=df,x="salary_in_usd",color="green")
plt.xlabel("Salaires (USD)") #titre axe x 
plt.ylabel("Fréquence") # titre axe y
plt.title ("Distribution des salaires en USD") #titre du graph
plt.show()#montre le graph
```

```python
# liste des pays concerné par l'étude 

print(df["company_location"].unique())

```

```python
#Salaire moyen par rôle
print( "\n 1 ###### Salaire moyen en USD par rôle ######\n")
Salaire_moyen_pays=df.groupby("job_title")["salary_in_usd"].sum()
display(Salaire_moyen_pays)

print( "\n 1 ###### Salaire moyen par rôle ######\n")
Salaire_moyen_pays=df.groupby("job_title")["salary"].sum()
display(Salaire_moyen_pays)

# Salaire moyen par pays
print( "\n 2 ###### Salaire Moyen en USD par pays ######\n")
Salaire_moyen_pays=df.groupby("employee_residence")["salary_in_usd"].sum()
display(Salaire_moyen_pays)

print( "\n 2 ###### Salaire Moyen par pays ######\n")
Salaire_moyen_pays=df.groupby("employee_residence")["salary"].sum()
display(Salaire_moyen_pays)

# Salaires moyens par niveau d'expérience
print( "\n 3 ###### Salaire Moyen par niveau ######\n")
Salaire_moyen_niveau=df.groupby("experience_level")["salary_in_usd"].mean()
display(Salaire_moyen_niveau)


print( "\n 3 ###### Salaire Moyen par niveau ######\n")
Salaire_moyen_niveau=df.groupby("experience_level")["salary_in_usd"].mean()
display(Salaire_moyen_niveau)
```

```python
# Analyse des corrélations
# Relation entre télétravail et salaire
corrélations_Pearson=df[["salary_in_usd","remote_ratio"]].corr() 
display(corrélations)
#Ici la corrélation est de 6% entre ces deux variable alors nous pouvons conclure qu'il n'y a pas de corrélation lineaire entre ces deux varaible 

sns.heatmap(corrélations)
```

#### Ici remote n'est pas une varible quantitaive continue mais discret donc essayons une autres methodes

```python
corrélations_Spearman=df[["salary_in_usd","remote_ratio"]].corr(method='spearman') #Correlation avec la methode Spearman
display(corrélations_Spearman)#toujours d'amelioration sur la corrélation 6%
#Representons les salaires en fonction du status remote sous format de boxplot
sns.boxplot(data=df,x="remote_ratio",y="salary_in_usd")
```

#### On remarque que la hauteurd des boxs 0 et 100 est casiment identiques se qui confirme notre theorie sur les faites que ces deux varibles ne sont pas correllés, Il reste la Box des travailleurs a temps partielle en teletravaille eux aurait tendance avec toucher moisn que les autres a tester en focntion des niveau dexperience dans chaque box 



```python

dummies = pd.get_dummies(df['remote_ratio'], prefix='remote') #Encodage de la variable discrete remote ration en troi categorie remplie de true or false pour chacune des tois 
df_corr_remote = pd.concat([df['salary_in_usd'], dummies], axis=1)# Concatenation du salaire avec ces trois nouvelle variables
corrélations_Pearson=df_corr_remote[["salary_in_usd","remote_0","remote_50","remote_100"]].corr() 
print(corrélations_Pearson)

sns.boxplot(data=df,x="remote_ratio",y="salary_in_usd",hue="experience_level")
```

#### Ici Nous avons confirmation sur la theorie que pour les employer en 50/50 on une correlation negative avec le salaire cela signifie qu'ils ont theoriquement plus de chance d'avoir un salaires inferieurs aux deux autres catégorie (correlation reste faible)

#### Car en ce qui concerne la correlation entre le salaire et le niveaux de d'experience cest plus on a d'experience plus le salaire augemente peux importe la situation de tele

```python
mapping_experience = {
    "EN": 1,
    "MI": 2,
    "SE":3,
    "EX":4,
}
df['experience_num'] = df['experience_level'].map(mapping_experience)
corrélations_Pearson_experience=df[["salary_in_usd","experience_num"]].corr() 
display(corrélations_Pearson_experience)
```

#### Nous obtenons belle est bien un correllations plus forte que celle du teletravailles et positives plus l'experience augmente plus le salaire augmente egalement  

```python
sns.regplot(data=df, x="experience_num", y="salary_in_usd", x_jitter=0.1, line_kws={"color": "red"})
plt.xticks([1, 2, 3, 4], ["EN", "MI", "SE", "EX"])
plt.show()
```
